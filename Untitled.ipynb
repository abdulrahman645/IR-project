```python
import pickle
```

```python
with open(f'./search/wikiren1ktraining_model.pkl', 'rb') as f:
    model = pickle.load(f)
with open(f'./search/wikiren1ktraining_speras_matrix.pkl', 'rb') as f:
    speras_matrix = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=2, random_state=0)
reduced_features = svd.fit_transform(speras_matrix)  # No need to call toarray()

reduced_cluster_centers = svd.transform(model.cluster_centers_)

plt.scatter(reduced_features[:,0], reduced_features[:,1], c=model.predict(speras_matrix))
plt.scatter(reduced_cluster_centers[:, 0], reduced_cluster_centers[:,1], marker='x', s=150, c='b')

plt.show()
```
